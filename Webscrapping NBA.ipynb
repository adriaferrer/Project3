```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from pprint import pprint
from lxml import html
from lxml.html import fromstring
import urllib.request
from urllib.request import urlopen
import random
import re
import scrapy
```

```python
url = 'https://hoopshype.com/salaries/players/'
```

```python
html = requests.get(url).content
html[0:600]
```

```python
soup = BeautifulSoup(html, "lxml")
soup
```

```python
table = soup.find_all('table',{'class':'hh-salaries-ranking-table hh-salaries-table-sortable responsive'})
table
```

```python
rows = table[0].find_all('tr')
header = rows[0].text.replace('\t','').strip().split('\n\n')
header
```

```python
players = rows[1:]
players_info = [player.text.replace('\t','').split('\n\n') for player in players]
players_info_clean = [player[3:] for player in players_info]
```

```python
for player in players_info_clean:
    player [1] = player[1].replace('\n','')
    player [6] = player[6].replace('\n','')
players_info_clean
```

```python
df_players_salaries = pd.DataFrame(players_info_clean)
df_players_salaries = df_players_salaries.rename(columns={0:'Full name',1:'2018/19',2:'2019/20',3:'2020/21',4:'2021/22',5:'2022/23',6:'2023/24'})
```
